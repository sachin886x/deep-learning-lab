```python
import torch
import numpy as np

# 1D
t1 = torch.tensor([1, 2, 3])
n1 = np.array([1, 2, 3])

# 2D
t2 = torch.tensor([[1, 2], [3, 4]])
n2 = np.array([[1, 2], [3, 4]])

# 3D
t3 = torch.randn(2, 3, 4)
n3 = np.random.randn(2, 3, 4)
```

```python
a = torch.tensor([1., 2., 3.])
b = torch.tensor([4., 5., 6.])

a + b        # addition
a - b        # subtraction
a * b        # multiplication
a / b        # division
```

```python
a_np = np.array([1., 2., 3.])
b_np = np.array([4., 5., 6.])

a_np + b_np
a_np - b_np
a_np * b_np
a_np / b_np
```

```python
# Dot product (1D)
torch.dot(a, b)
np.dot(a_np, b_np)

# Matrix multiplication
A = torch.randn(2, 3)
B = torch.randn(3, 4)
torch.matmul(A, B)      # or A @ B

A_np = np.random.randn(2, 3)
B_np = np.random.randn(3, 4)
A_np @ B_np
```

```python
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

x[0]          # first row
x[:, 1]       # second column
x[0, 1]       # element

# Boolean masking
mask = x > 3
x[mask]       # elements > 3

# Subtensor
x[:2, :2]
```

```python
x_np = np.array([[1, 2, 3],
                 [4, 5, 6]])
x_np[x_np > 3]
```

```python
t = torch.arange(6)

t.view(2, 3)          # only works if contiguous
t.reshape(2, 3)      # safer

t.unsqueeze(0)        # add dim at axis 0 -> (1, 6)
t.unsqueeze(1)        # -> (6, 1)

t2 = torch.tensor([[1, 2, 3]])
t2.squeeze()          # removes size-1 dims
```

```python
n = np.arange(6)
n.reshape(2, 3)       # same idea, always safe
```

```python
A = torch.randn(3, 4)
b = torch.randn(4)

A + b     # b broadcast to (3, 4)
```

```python
A_np = np.random.randn(3, 4)
b_np = np.random.randn(4)
A_np + b_np
```
